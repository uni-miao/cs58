```python
import pandas as pd

df = pd.read_csv("dataset_14_08_2025_completion.csv")

retraction_cols = ["RetractionDOI", "RetractionPubMedID"]
original_cols = ["OriginalPaperDOI", "OriginalPaperPubMedID"]

def extract_identifiers(columns):
    identifiers = []
    seen = set()  
    for col in columns:
        values = df[col].values.flatten()
        for x in values:

            str_x = str(x).strip().lower()
            
            if pd.notna(x) and str_x not in ["", "0", "unavailable"]:
                if "PubMedID" in col:
                    processed = str(int(float(x)))
                else:  
                    processed = str(x).strip()
                
                if processed not in seen:
                    seen.add(processed)
                    identifiers.append(processed)
    return identifiers

retraction_identifiers = extract_identifiers(retraction_cols)

original_identifiers = extract_identifiers(original_cols)

with open("Identifiers_Retraction.txt", "w") as f:
    f.write("\n".join(retraction_identifiers))

with open("Identifiers_Original.txt", "w") as f:
    f.write("\n".join(original_identifiers))
    
```
